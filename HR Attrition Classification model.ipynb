```python
cat = []
num = []


for i in df_resampled.columns:
    if df_resampled[i].dtype == 'O':
        cat.append(i)
    else:
        num.append(i)
```

```python
cat
```

```python
num
```

```python
# Cat 

for i in cat:
    vc = df_resampled[i].value_counts()
    plt.bar(vc.index, height = vc , color = 'b')
    plt.xlabel(i)
    plt.xticks(rotation= 90)
    plt.ylabel('Count')
    plt.show()
```

```python
# Numerical

for i in num:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(df_resampled[i],color = 'c')
    ax[1].boxplot(df_resampled[i])
    plt.suptitle(i)
    plt.show()
```

```python
# iqr winsorization --
# replace lower outliers with lower fence
# replace upper outliers with upper fence


for i in num:
    q1 = df_resampled[i].quantile(0.25)
    q3 = df_resampled[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    df_resampled.loc[df_resampled[i]<lf, i] = lf
    df_resampled.loc[df_resampled[i]>uf, i] = uf
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, confusion_matrix

# Training the model
model = LogisticRegression()
model.fit(x_train, y_train)

# Predicting on training and testing data
y_train_predicted = model.predict(x_train)
y_test_predicted = model.predict(x_test)

# Training Performance
print("Training Performance: ")
print("Training accuracy: ", accuracy_score(y_train, y_train_predicted))
print("Training precision: ", precision_score(y_train, y_train_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Training recall: ", recall_score(y_train, y_train_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_train, y_train_predicted))
print("Training classification report: ", classification_report(y_train, y_train_predicted))

# Testing Performance
print("Testing Performance: ")
print("Testing accuracy: ", accuracy_score(y_test, y_test_predicted))
print("Testing precision: ", precision_score(y_test, y_test_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Testing recall: ", recall_score(y_test, y_test_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_test, y_test_predicted))
print("Testing classification report: ", classification_report(y_test, y_test_predicted))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the parameter grid to search
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  # Number of neighbors to test
    'weights': ['uniform', 'distance'],  # Weight function used in prediction
    'metric': ['euclidean', 'manhattan', 'chebyshev']  # Distance metrics
}

# Initialize the KNN model
knn_model = KNeighborsClassifier()

# Initialize the GridSearchCV with 5-fold cross-validation
grid_search = GridSearchCV(estimator=knn_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)

# Fit GridSearchCV on the training data
grid_search.fit(x_train, y_train)

# Get the best parameters from GridSearchCV
best_params = grid_search.best_params_
print(f"Best Hyperparameters: {best_params}")

# Get the best model from GridSearchCV
best_knn_model = grid_search.best_estimator_

# Predict using the best model
y_train_predicted = best_knn_model.predict(x_train)
y_test_predicted = best_knn_model.predict(x_test)

# Evaluate the model performance on training data
print("Training Performance: ")
print("Training accuracy: ", accuracy_score(y_train, y_train_predicted))
print("Training classification report: ", classification_report(y_train, y_train_predicted))
print(confusion_matrix(y_train, y_train_predicted))

# Evaluate the model performance on testing data
print("Testing Performance: ")
print("Testing accuracy: ", accuracy_score(y_test, y_test_predicted))
print("Testing classification report: ", classification_report(y_test, y_test_predicted))
print(confusion_matrix(y_test, y_test_predicted))
```

# Gric Search CV - Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the Decision Tree Classifier
dtree = DecisionTreeClassifier()

# Define the parameter grid to search over
param_grid = {
    'max_depth': [3, 5, 10,15,20, None],  # Depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples required at a leaf node
    'criterion': ['gini', 'entropy']  # Split criterion
}

# Initialize the GridSearchCV
grid_search = GridSearchCV(estimator=dtree, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)

# Fit the GridSearchCV to the training data
grid_search.fit(x_train, y_train)

# Print the best hyperparameters found
print(f"Best Hyperparameters: {grid_search.best_params_}")

# Get the best model from GridSearchCV
best_dtree_model = grid_search.best_estimator_

# Predict on training data using the best model
y_train_predicted = best_dtree_model.predict(x_train)

# Evaluate on training data
print("\nTraining Model Performance:")
print("Training Accuracy:", accuracy_score(y_train, y_train_predicted))
print("Training Classification Report:\n", classification_report(y_train, y_train_predicted))
print("Training Confusion Matrix:\n", confusion_matrix(y_train, y_train_predicted))

# Predict on testing data using the best model
y_test_predicted = best_dtree_model.predict(x_test)

# Evaluate on testing data
print("\nTesting Model Performance:")
print("Testing Accuracy:", accuracy_score(y_test, y_test_predicted))
print("Testing Classification Report:\n", classification_report(y_test, y_test_predicted))
print("Testing Confusion Matrix:\n", confusion_matrix(y_test, y_test_predicted))
```
