```python
# Under sampling

no_cat = df[df['Attrition'] == 'No'].sample(711)
no_cat
```

```python
yes_cat = df[df['Attrition'] == 'Yes']
yes_cat
```

```python
df_resampled = pd.concat([no_cat,yes_cat], axis = 0)
df_resampled
```

```python
df_resampled.drop('Over18', axis = 1, inplace = True)
```

```python
cat = []
num = []


for i in df_resampled.columns:
    if df_resampled[i].dtype == 'O':
        cat.append(i)
    else:
        num.append(i)
```

```python
cat
```

```python
num
```

```python
# Cat 

for i in cat:
    vc = df_resampled[i].value_counts()
    plt.bar(vc.index, height = vc , color = 'b')
    plt.xlabel(i)
    plt.xticks(rotation= 90)
    plt.ylabel('Count')
    plt.show()
```

```python
# Numerical

for i in num:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(df_resampled[i],color = 'c')
    ax[1].boxplot(df_resampled[i])
    plt.suptitle(i)
    plt.show()
```

```python
# iqr winsorization --
# replace lower outliers with lower fence
# replace upper outliers with upper fence


for i in num:
    q1 = df_resampled[i].quantile(0.25)
    q3 = df_resampled[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    df_resampled.loc[df_resampled[i]<lf, i] = lf
    df_resampled.loc[df_resampled[i]>uf, i] = uf
```

```python
X_df = df_resampled.drop(['Attrition','Department','JobRole','EducationField','BusinessTravel','MaritalStatus'], axis = 1)
y = df_resampled['Attrition']
X_df
```

```python
y
```

```python
X = X_df.values
```

```python
Y = y.values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
y_train_predicted = model.predict(x_train)
y_test_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report ,accuracy_score,precision_score,recall_score,confusion_matrix
```

```python
print("Model Performance: ")
print("Training accuracy: ",accuracy_score(y_train,y_train_predicted))
print("Training precision: ", precision_score(y_train, y_train_predicted))
print("Training recall: ", recall_score(y_train, y_train_predicted))
print(confusion_matrix(y_train, y_train_predicted))
print("Training classification report: ",classification_report(y_train, y_train_predicted))
```

```python
# Test 

#  Accuracy > 80% 

print("Model Performance: ")
print("Testing accuracy: ",accuracy_score(y_test,y_test_predicted))
print("Testing precision: ", precision_score(y_test,y_test_predicted))
print("Testing recall: ", recall_score(y_test,y_test_predicted))
print(confusion_matrix(y_test,y_test_predicted))
print("Testing classification report: ",classification_report(y_test,y_test_predicted))

```
